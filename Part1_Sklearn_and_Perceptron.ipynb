```python
# Imports

from sklearn.linear_model import Perceptron
from sklearn.preprocessing import MinMaxScaler
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
# Load the digits dataset and look at a few examples

digits = load_digits()
plt.gray()

for i in range(10):
    print('Digit = %d' % i)
    plt.matshow(digits.images[i])
    plt.show()
```

```python
# Get the attribute/feature vectors and the class labels

X = digits.data
y = digits.target
```

```python
# Convert 10 digit problem to one in which the goal is to tell if a digit is
# 0 or not 0, i.e. convert to binary classification

y = [1 if a == 0 else 0 for a in y]
```

## Task 1: Read about scikit's train/test splitter

https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html

```python
# Split the data into a training set and a test set

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.5,
                                                    random_state=42)
```
